|               | pred dog   | pred cat   |
|:------------  |-----------:|-----------:|
| actual dog    |         46 |         7  |
| actual cat    |         13 |         34 |


* In the context of this problem, what is a false positive?
* In the context of this problem, what is a false negative?
* How would you describe this model?


```python
# false positive = predict + , actural - ;

# 1. pred is dog + actual is cat
# 2. pred is cat + actual is dog
```

```python
# false nefative = predict - , actural + ;

# 1. pred is cat + actual is dog
# 2. pred is dog + actual is cat
```

```python
# How would you describe this model?
```

```python
data = {'pred_dog':[46, 13], 'pred_cat':[7, 34]} 
  
# Creates pandas DataFrame. 
df = pd.DataFrame(data, index =['actual_dog', 'actual_cat']) 

df
```

#### positive : dog

* TP : pred dog + actual dog = 46
* TN : pred not dog + actual not dog = 34
* FP : pred dog + actual not dog = 13
* FN : pred not dog + actual dog = 7

```python
# accuracy: (TP + TN) / (TP + TN + FP + FN)
(46+34) / (46+34+13+7)
```

```python
# precision: TP / (TP + FP)
46 / (46+13)
```

```python
# recall: TP / (TP + FN)
46 / (46+7)
```

#### positive: cat

```python
# A false positive is precting a cat when the observation is a dog
# A false negative is predicting a dog when it's actually a cat (a miss)

tp= 34
tn = 46
fp = 7
fn = 13

accuracy = (tp + tn) / (tp +tn +fp + fn)
recall = tp / (tp + fn)
precision = tp / (tp + fp)

print('Cat-Classifier')
print(f'Accuracy: {accuracy:.2%}')
print(f'Recall: {recall:.2%}')
print(f'Precision: {precision:.2%}')
```

### Question 1
#### metric : Recall (minimizing FN) :  TP / (TP+FN)

```python
# positive = defect

# Which model would be the best fit for this use case? --- Model_3
subset = df_c3[df_c3.actual == 'Defect']
subset
```
