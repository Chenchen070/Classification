```python
import os

def get_titanic_data():
    filename = "titanic.csv"
    
    if os.path.isfile(filename):
        return pd.read_csv(filename, index_col=[0])
    else:
        df = new_titanic_data()
        df.to_csv(filename)
        return df 
```

```python
import os

def get_titanic_data():
    filename = "titanic.csv"
    
    # if file is available locally, read it
    if os.path.isfile(filename):
        return pd.read_csv(filename)
    
    # if file not available locally, acquire data from SQL database
    # and write it as csv locally for future use
    else:
        # read the SQL query into a dataframe
        df = new_titanic_data()
        
        # Write that dataframe to disk for later. Called "caching" the data for later.
        df.to_csv(filename)

        # Return the dataframe to the calling code
        return df 
```

```python
df_titanic = get_titanic_data()
df_titanic.head()
```

```python
def get_iris_data():
    filename = "iris.csv"
    
    if os.path.isfile(filename):
        return pd.read_csv(filename, index_col=[0])
    else:
        df = new_iris_data()
        df.to_csv(filename)
        return df 
```

#### 3. Make a function named get_telco_data that returns the data from the telco_churn database in SQL. In your SQL, be sure to join all 4 tables together, so that the resulting dataframe contains all the contract, payment, and internet service options. Obtain your data from the Codeup Data Science Database.

```python
def new_telco_data():
    return pd.read_sql('select * from customers left join internet_service_types using(internet_service_type_id) left join payment_types using (payment_type_id) left join contract_types using (contract_type_id)',
                      get_connection('telco_churn'))
new_telco_data().head()
```

```python
def get_telco_data():
    filename = "telco.csv"
    
    if os.path.isfile(filename):
        return pd.read_csv(filename, index_col=[0])
    else:
        df = new_telco_data()
        df.to_csv(filename)
        return df 
```

### Using the Iris Data:

1. Use the function defined in acquire.py to load the iris data.

2. Drop the species_id and measurement_id columns.

3. Rename the species_name column to just species.

4. Create dummy variables of the species name and concatenate onto the iris dataframe. (This is for practice, we don't always have to encode the target, but if we used species as a feature, we would need to encode it).

5. Create a function named prep_iris that accepts the untransformed iris data, and returns the data with the transformations above applied.

```python
import acquire
```

```python
df = acquire.get_iris_data()
df.head()
```

```python
df.info()
```

```python
# Drop the species_id and measurement_id columns.
df = df.drop(columns = ['species_id', 'measurement_id'])
df.head()
```

```python
# Rename the species_name column to just species.
df = df.rename(columns = {'species_name': 'species'})
df.head()
```

```python
# Create dummy variables of the species name and concatenate onto the iris dataframe. (This is for practice, we don't always have to encode the target, but if we used species as a feature, we would need to encode it).
dummy_df = pd.get_dummies(df[['species']], dummy_na= False, drop_first=True)
dummy_df.head()
```

```python
# concatenate
df = pd.concat([df, dummy_df], axis = 1)
df.head()
```

```python
# Create a function named prep_iris that accepts the untransformed iris data, and returns the data with the transformations above applied.

def prep_iris(df):
    df = df.drop(columns = ['species_id', 'measurement_id'])
    df = df.rename(columns = {'species_name': 'species'})
    dummy_df = pd.get_dummies(df[['species']], dummy_na= False, drop_first=True)
    df = pd.concat([df, dummy_df], axis = 1)
    
    return df
    
```

```python
df1 = acquire.get_iris_data()
df1.head()
```

```python
df1 = prep_iris(df1)
df1.head()
```

```python
# Create a function named prep_telco that accepts the raw telco data, and returns the data with the transformations above applied.

def prep_telco(df):
    df["total_charges"] = pd.to_numeric(df["total_charges"], errors="coerce")
    df = df.drop_duplicates()
    df = df.dropna()
    cols_to_drop = ['contract_type_id', 'payment_type_id', 'internet_service_type_id', 'online_security', 'online_backup', 'device_protection', 'tech_support', 'streaming_tv','streaming_movies', 'paperless_billing']
    df = df.drop(columns = cols_to_drop)
    df_dummy = pd.get_dummies(df[['gender', 'partner', 'dependents', 'phone_service', 'multiple_lines', 'churn']], dummy_na=False, drop_first=[True,True,True,True,True,True])
    df = pd.concat([df, df_dummy],axis=1)
    return df
```
